```python

# Reza Ashour Farsiani 2
# diffusion in ID

import numpy as np
import matplotlib.pyplot as plt

x=np.linspace (0, 1, 101)
u=np.zeros(101)
u[50]=1
for time in range(100):
    u[1:-1] += 0.5*(u[2:] - 2*u[1:-1] + u[:-2])
    
plt.plot(x,u,'r-')
plt.show()
```

```python

# Reza Ashour Farsiani 1
# python script for diffusion in ID
# equation du/dt=d^2u/dt

import numpy as np
import matplotlib.pyplot as plt

x=np.linspace (0, 1, 101)
u=np.zeros(101)
u[50]=1

for time in range(100):
    for i in range (1, 100):
        u[i] += (u[i+1] - 2*u[i] + u[i-1])/4.
        
plt.plot(x,u,'g-')
plt.xlabel('along the lenght of the rod')
plt.ylabel('temperature')
plt.show()
```

```python
#3.

# Reza Ashour Farsiani
# python script for diffusion in ID
# equation du/dt=d^2u/dt

import numpy as np
import matplotlib.pyplot as plt

x=np.linspace (0, 1, 101)
u=np.zeros(101)
u[0], u[100] = 1, 1
for time in range(100):
    for i in range (1, 100):
        u[i] += (u[i+1] - 2*u[i] + u[i-1])/4.
        
plt.plot(x,u,'b--')
plt.xlabel('along the lenght of the rod')
plt.ylabel('temperature')
plt.show()
```

```python
#4.

# Reza Ashour Farsiani
# uniformly heated rod

T0, L=1, 1
x=np.linspace (0, L, 101)
u= T0*np.ones (101)
u[0], u[100] = 0, 0

for time in range(100):
    u[1:-1] += 0.25*(u[2:] - 2*u[1:-1] + u[:-2])
    
t=1/300
f=lambda m: 2/(2*m-1)*2*T0/np.pi*np.exp(-((2*m-1)*np.pi/L)**2*t)*np.sin((2*m-1)*np.pi/L*x)
exact = sum([f(m) for m in range(1, 20)])

plt.plot(x, u, 'o', x, exact, lw=3)
plt.xlabel('lenght of the rod')
plt.ylabel('initial temperature')
```

```python
#5.

# Reza Ashour Farsiani
#The temperature of a point changes with time

import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0, 1, 101)
u=np.zeros(101)
u[50]=1

u0=[]
T=range(100)

for time in T:
    for i in range(1, 100):
        u[i] += (u[i+1]-2*u[i]+u[i-1])/4.
    u0.append(u[50])
    
plt.plot(T, u0, 'y', lw=3)
plt.xlabel('time')
plt.ylabel('temp at heating point')
plt.show()
```

```python
# Reza Ashour Farsiani 7

import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0, 1, 101)
y=np.linspace(0, 1, 101)
u=np.zeros((101, 101))
u[50, 50]=1

for time in range(200):
    for i in range(1, 100):
        for j in range(1,100):
            u[i, j] += (u[i+1, j] + u[i, j+1] + u[i-1, j] + u[i, j-1] - 4*u[i, j-1])/4.
            

import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits import mplot3d
X, T = np.meshgrid(x, t)
fig = plt.figure()
ax=plt.axes(projection='3d')
ax.plot_surface(X, T, u, cmap=cm.coolwarm)
plt.show()
```
