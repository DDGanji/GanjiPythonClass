```python

# Reza Ashour Farsiani
# python script for diffusion in ID
# equation du/dt=d^2u/dt

import numpy as np
import matplotlib.pyplot as plt

x=np.linspace (0, 1, 101)
u=np.zeros(101)
u[50]=1

for time in range(100):
    for i in range (1, 100):
        u[i] += (u[i+1] - 2*u[i] + u[i-1])/4.
        
plt.plot(x,u,'g-')
plt.xlabel('along the lenght of the rod')
plt.ylabel('temperature')
plt.show()
```

```python

# Reza Ashour Farsiani
# diffusion in ID

import numpy as np
import matplotlib.pyplot as plt

x=np.linspace (0, 1, 101)
u=np.zeros(101)
u[50]=1
for time in range(100):
    u[1:-1] += 0.25*(u[2:] - 2*u[1:-1] + u[:-2])
    
plt.plot(x,u,'r-')
plt.show()
```
